# Image Segmentation using cv2 and KMeans Clustering algorithm

This exercise aims at image segmentation using KMeans clustering algorithm where the same image is segmented to occupy less memeory space.

```python
import cv2
from PIL import Image
from numpy import asarray
```

```python
img = Image.open("tiger.png")
```

```python
img.show()
```

```python
numpydata = asarray(img)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85)
 
# then perform k-means clustering with  number of clusters = 3
#also random centres are initially chosen for k-means clustering
k = 16
retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria,3, cv2.KMEANS_RANDOM_CENTERS)

  
# convert data into 8-bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]
 
# reshape data into the original image dimensions
segmented_image = segmented_data.reshape((numpydata.shape))
 
plt.imshow(segmented_image)
```
