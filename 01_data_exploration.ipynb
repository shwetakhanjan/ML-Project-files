```python
import numpy as np
import pandas as pd
import os

# --- Set number of students ---
n_students = 500

# --- Generate synthetic features ---
np.random.seed(42)
student_id = [f"S{i:03d}" for i in range(1, n_students+1)]
age = np.random.randint(10, 15, n_students)
grade = np.random.randint(5, 9, n_students)
homework_completion = np.random.normal(85, 10, n_students).clip(50, 100)
practice_tests_avg = np.random.normal(75, 15, n_students).clip(40, 100)
class_participation = np.random.normal(7, 2, n_students).clip(0, 10)
geometry_score = np.random.normal(70, 15, n_students).clip(30, 100)
prealgebra_score = np.random.normal(72, 12, n_students).clip(40, 100)
recent_quiz_avg = np.random.normal(70, 15, n_students).clip(30, 100)
study_hours_per_week = np.random.normal(5, 2, n_students).clip(0, 10)
at_risk = ((homework_completion < 60) |
           (practice_tests_avg < 60) |
           (geometry_score < 50)).astype(int)

df = pd.DataFrame({
    "student_id": student_id,
    "age": age,
    "grade": grade,
    "homework_completion": homework_completion,
    "practice_tests_avg": practice_tests_avg,
    "class_participation": class_participation,
    "geometry_score": geometry_score,
    "prealgebra_score": prealgebra_score,
    "recent_quiz_avg": recent_quiz_avg,
    "study_hours_per_week": study_hours_per_week,
    "at_risk": at_risk
})

# --- Save CSV inside repo ---
# Get current working directory (where the notebook is)
cwd = os.getcwd()
print("Current working directory:", cwd)

# Define path relative to repo root
save_path = os.path.join(cwd, "data", "processed", "student_math_synthetic.csv")

# Make sure folder exists
os.makedirs(os.path.dirname(save_path), exist_ok=True)

# Save CSV
df.to_csv(save_path, index=False)
print(f"CSV saved at: {save_path}")
```
